```python
import pandas as pd

# Load the Excel file
file_path = '/Users/xiaofanjiao/Desktop/vegan_cleaned.xlsx'
# Load the data from 'vegan' sheet
data = pd.read_excel(file_path, header=None)
data.head(20)  # Display the first 20 rows to see the data structure.
```

```python
# Removing duplicates where all column values are the same
data_cleaned = data.drop_duplicates()

# Checking how many rows were removed
rows_removed = data.shape[0] - data_cleaned.shape[0]

data_cleaned.head(), rows_removed
```

```python
# Save the structured data to a new Excel file
output_file_path = '/Users/xiaofanjiao/Desktop/vegan_finished.xlsx'
data_cleaned.to_excel(output_file_path, index=False)
```
