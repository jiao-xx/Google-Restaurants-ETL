```python
import pandas as pd

# Load the Excel file
file_path = '/Users/xiaofanjiao/Desktop/vegan.xlsx'
```

```python
# Load the data from 'vegan' sheet
vegan_data = pd.read_excel(file_path, sheet_name='Vegan', header=None)
vegan_data.head(20)  # Display the first 20 rows to see the data structure.
```

```python
import pandas as pd

def parse_vegan_data(data):
    # Initialize an empty list to hold the parsed data
    parsed_data = []
    
    # Temporary dictionary to store each entry
    entry = {}
    
    # Iterate over the rows in the dataframe
    for index, row in data.iterrows():
        # Convert the cell content to string and check for '------'
        if isinstance(row[0], str) and '------' in row[0]:
            # If the entry dictionary is not empty, append it to the list
            if entry:
                parsed_data.append(entry)
                entry = {}
        else:
            # Check if row is not empty and contains ': '
            if isinstance(row[0], str) and ': ' in row[0]:
                key, value = row[0].split(': ', 1)
                entry[key] = value
            
    # Append the last entry if it's not empty
    if entry:
        parsed_data.append(entry)
    
    # Create a dataframe from the parsed data
    structured_data = pd.DataFrame(parsed_data)
    
    # Return the structured dataframe
    return structured_data

# Assuming 'vegan_data' is your initial DataFrame
structured_vegan_data = parse_vegan_data(vegan_data)

# Display the structured dataframe
structured_vegan_data.head()
```

```python
# Save the structured data to a new Excel file
output_file_path = '/Users/xiaofanjiao/Desktop/vegan_cleaned.xlsx'
structured_vegan_data.to_excel(output_file_path, index=False)
```
